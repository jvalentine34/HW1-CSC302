Homework 1: Use pandas on cause_of_death.cvs

A

```python
deaths = pd.read_csv('/content/drive/MyDrive/DATA/causes_of_death.csv')
print(deaths.info, '\n')

print('Size =', deaths.size, '\n')

clms = deaths.shape[1]
print('# of Columns =', clms, '\n')

rws = deaths.index
number_of_rows = len(rws)
print('# of Rows =', number_of_rows, '\n')
```

B.

```python
dataType = deaths.dtypes
print('The data type of each column:')
print(dataType)
```

```python
print('Average number of deaths =', deaths['Deaths'].mean())
```

C.

```python
deaths.groupby(['State'])['Deaths'].sum()
```

D.

```python
deaths.groupby(['State'])['Deaths'].mean()
```

E.

```python
deaths.groupby(['State','Gender'])['Deaths'].sum()
```

```python
deaths.dropna(subset = ['State','Gender'])
deaths.head()
```

```python
deaths['Ratio'] = (deaths['Population'] / deaths['Deaths'].sum()) * 100
deaths.head()
```
